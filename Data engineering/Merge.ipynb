```python
import pandas as pd  
```

```python
## open product CSV
pd_reader =pd.read_csv(open("olist_products_dataset3.csv"))

products_dataset = pd_reader
```

```python
products_dataset.info()
```

```python
products_dataset.isnull().sum()
```

```python
## fill the gap 

products_dataset["product_category_name"].fillna("other", inplace = True)
```

```python
products_dataset.isnull().sum()
```

```python
products_dataset["product_category_name"]
```

```python
## change the other type of products become 'other' product
for col in products_dataset["product_category_name"]:
    if col != 'cama_mesa_banho'and col != 'esporte_lazer'and col != 'beleza_saude' and col !='moveis_decoracao' and col != 'utilidades_domesticas'and col != 'automotivo'and col != 'brinquedos'and col != 'informatica_acessorios'and col != 'relogios_presentes'and col != 'telefonia':
        products_dataset["product_category_name"].replace(col,'other',inplace=True)
```

```python
## categorize product type
new_products_dataset = pd.get_dummies(products_dataset, columns=["product_category_name"], prefix=["type_"])
new_products_dataset
```

```python
## drop some features that we don't need
new_products_dataset=new_products_dataset.drop(columns=['product_name_lenght','product_weight_g','product_length_cm','product_height_cm','product_width_cm'])
```

```python
## introduce another items CSV 

pd_reader2 =pd.read_csv(open("order_items_edited.csv"))
```

```python
OrderItems_Dataset = pd_reader2
```

```python
OrderItems_Dataset.head()
```

```python
New_Dataset = pd.merge(OrderItems_Dataset,new_products_dataset,how='left',left_on='product_id',right_on='product_id')
```

```python
New_Dataset.info()
```

```python
New_Dataset=New_Dataset.drop(columns=['seller_id'])
```
