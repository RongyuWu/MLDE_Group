```python
import pandas as pd
io = r'D:\ebm\machine learning\brazilian-ecommerce\olist_orders_dataset.csv'
order_data = pd.read_csv(io)
order_data.head()
```

```python
order_data.info()
```

```python
order_data.isnull().sum()
```

```python
order_data['order_purchase_timestamp'] = pd.to_datetime(order_data['order_purchase_timestamp'])
```

```python
order_data.info()
```

```python
order_data['order_delivered_customer_date'] = pd.to_datetime(order_data['order_delivered_customer_date'])
order_data['order_estimated_delivery_date'] = pd.to_datetime(order_data['order_estimated_delivery_date'])
```

```python
order_data.info()
```

```python
order_data.info() #object是pandas里的string类型
```

```python
order_data = pd.get_dummies(order_data, columns=["order_status"], prefix=["status_"])
order_data.head()
```

```python
order_data['period'] = order_data['order_estimated_delivery_date']-order_data['order_purchase_timestamp']
```

```python
order_data.head()
```

```python
order_data['ship_score'] = (order_data['order_delivered_customer_date']-order_data['order_purchase_timestamp'])/order_data['period']
```

```python
order_data.head()
```

```python
order_data.to_csv('order_pd.csv')
```
